```python
#UA Innovate Hackathon 3/30/2015
#Mae Olsen, Casey Derringer, Liam Fitzpatrick

import numpy as np
import matplotlib.pyplot as plt

xVals = range(1, 21) # number of coordinators to test
yVals = np.zeros(20)  # number of hours between task completion and closure
yVals[0] = 604  # annual sum hours between task completion and closure

for i in range(1,19):
  yVals[i] = yVals[i-1]/2 # assume that each full-time additional coordinator cuts work in half

xVals = np.array(xVals)
def exp_decay(x, a, b): # exponential decay function
  return a*np.exp(-b*(x -1))

y_fitted = exp_decay(xVals, 604, np.log(2)) # calculated y values

#Curve Plotting
plt.plot(xVals, yVals, 'ro')
plt.plot(xVals, y_fitted, 'b')
plt.xlabel('Number of Employees')
plt.ylabel('Closed Time - Completed Time')
plt.xticks(range(0, 21, 1))
plt.title('Number of Coordinators vs. Time Difference Between Closure and Completion')
plt.legend(['Raw Data Points', 'Fitted Curve: 640*e^(-log(2)(x-1))'])
plt.show()


mse = np.mean((yVals-y_fitted)**2)
print("Mean Squared Error: %e" %(mse))
```

```python
import sympy
from scipy.misc import derivative

def fun(x): # function redefinition
  return 604*np.exp(-np.log(2)*(x-1))

def diff(x):  #first derivative
  return derivative(fun, x)

def diff2(x): #second derivative
  return diff(diff(x))

#Curve Plotting
plt.plot(xVals, diff(xVals))
plt.title('First Derivative Graph')
plt.show
```

```python
# Curve Plotting
plt.plot(xVals, diff2(xVals))
plt.plot([2]*20, diff2(xVals))
plt.title('Second Derivative Graph')
plt.legend(['Second Derivative', 'x = 2'])
#intersection at 2 implies optimal number of
#coordinators = 2
plt.show()
```
