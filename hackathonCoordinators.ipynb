```python
#UA Innovate Hackathon 3/30/2025
#Mae Olsen, Casey Derringer, Liam Fitzpatrick

import numpy as np
import matplotlib.pyplot as plt

xVals = range(1, 40) # number of coordinators to test
xVals = np.array(xVals)

# hyperbolic decay function
# 604 hour delay would decrease with added employees
def hyper_decay(x):
  return 604/(x)

yVals = hyper_decay(xVals) # number of hours between task completion and closure

y_prime = np.gradient(yVals) # first derivative of hyper_decay

y_doublePrime = np.gradient(y_prime) # second derivative of hyper_decay

#Curve Plotting
plt.plot(xVals, yVals, 'b')
plt.plot(xVals, y_prime)
plt.plot(xVals, y_doublePrime)
plt.plot([4]*39, range(-300, 636, 24)) # y = 4
plt.plot([2]*39, range(-300, 636, 24)) # y = 2
plt.plot([3]*39, range(-300, 636, 24)) # y = 3
plt.plot([5]*39, range(-300, 636, 24)) # y = 5
plt.plot([8]*39, range(-300, 636, 24)) # y = 8
plt.xticks(range(0, 41, 2))
plt.title('Comparison of Function and Derviatives')
plt.xlabel('Number of Coordinators')
plt.legend(['f(x)=640/x', "f'(x)", 'f"(x)', 'y=4', 'y=2', 'y=3', 'y=5', 'y=8'])
plt.show()

# if you wanted to find a concrete threshold for how close you want the second derivative to
# equal zero, you could adjust this value
threshold = 1
index = np.where(np.abs(y_doublePrime)< threshold)[0][0]

# Several points show slight breaking where the slope decreases, decaying at a slower rate.
# Since there is no distinct elbow in a hyperbolic decay, the optimal number of employees lies
# in this range of values and depends on how efficient the employer wants to be based on this information.
```
